```python
import pandas as pd
import json
```

```python
# 读取 CSV 文件
file_path = 'pathformer.csv'
data = pd.read_csv(file_path)

# 去除 file_name 列中的 ".csv" 后缀
data['file_name'] = data['file_name'].str.replace('.csv', '', regex=False)

# 从 strategy_args 列中提取出 horizon 的值，并创建新的 horizon 列
data['horizon'] = data['strategy_args'].apply(lambda x: json.loads(x).get('horizon'))

# 更新 file_name 列，将其重写为 file_name-horizon，并替换原有的 "-" 为 "_"
data['file_name'] = data['file_name'].str.replace('-', '_', regex=False) + '-' + data['horizon'].astype(str)

# 只保留 mae, mse 和 file_name 三列
data_filtered = data[['mae', 'mse', 'file_name']]

# 将每一行拆分为两行，一行为 mae，一行为 mse
mae_rows = data_filtered[['file_name', 'mae']].copy()
mae_rows['file_name'] = mae_rows['file_name'] + '-mae'
mae_rows = mae_rows.rename(columns={'mae': 'value'})

mse_rows = data_filtered[['file_name', 'mse']].copy()
mse_rows['file_name'] = mse_rows['file_name'] + '-mse'
mse_rows = mse_rows.rename(columns={'mse': 'value'})

# 合并 mae 和 mse 的行
reshaped_data = pd.concat([mae_rows, mse_rows], ignore_index=True)
reshaped_data
```
